Data Cleaning

```python
#Remove 'Cr'& convert into Decimal value
housing_df['Ticket_Price_Cr']=housing_df['Ticket_Price_Cr'].astype(str).str.replace(' Cr','', regex=False).str.replace('₹','',regex=False).astype(float)
housing_df['Ticket_Price_Cr']
```

```python
#Filling Missing Values
housing_df['Unit_Size_Sqft'] = housing_df['Unit_Size_Sqft'].fillna(housing_df['Unit_Size_Sqft'].mean())
housing_df['Ticket_Price_Cr'] = housing_df['Ticket_Price_Cr'].fillna(housing_df['Ticket_Price_Cr'].mean())
housing_df['Amenity_Score'] = housing_df['Amenity_Score'].fillna(housing_df['Amenity_Score'].median())
housing_df['Buyer_Comments'] = housing_df['Buyer_Comments'].fillna("No Comments")
housing_df
```

```python
#Normalizing Columns 
columns = ['Micro_Market', 'Developer_Name', 'Configuration','Transaction_Type', 'Buyer_Type', 'Possession_Status',
             'Sales_Channel', 'NRI_Buyer']
```

```python
for col in columns:
    housing_df[col] = housing_df[col].str.strip().str.title()
    
```

```python
housing_df.dtypes
```

```python
housing_df
```

```python
housing_df['Purchase_Quarter'] = pd.to_datetime(housing_df['Purchase_Quarter'])
housing_df['Quarter_Number'] = housing_df['Purchase_Quarter'].dt.quarter
housing_df['Quarter_Number']
```

```python
print(housing_df['Purchase_Quarter'].unique())
```

```python
housing_df['Configuration']=housing_df['Configuration'].astype(str).str.upper()
```

```python
housing_df['Ticket_Price_Cr'] = pd.to_numeric(housing_df['Ticket_Price_Cr'], errors='coerce')
housing_df['Ticket_Price_Cr'] = housing_df['Ticket_Price_Cr'].astype('float64')
housing_df['Price_per_sqft']=housing_df['Ticket_Price_Cr']*10000000 / housing_df['Unit_Size_Sqft']
housing_df
```

```python
housing_df['Booking_Flag'] = housing_df['Transaction_Type'].str.lower().map({'primary': 1, 'secondary': 0})
housing_df['Booking_Flag']
```

```python
housing_df.rename(columns={"Developer_Name": "Builder"}, inplace=True)
```
